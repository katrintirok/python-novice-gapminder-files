# Making plots with matplotlib

```python
# use this command before starting with any plotting to make sure plots appear 'inline'
%matplotlib inline   
```

```python
import matplotlib.pyplot as plt  # import sublibrary pyplot from matplotlib using alias plt
```

```python
time = [1, 2, 3, 4]  # x-values
position = [0, 100, 200, 300]   # y-values
```

```python
plt.plot(time, position)  # basic plot of position vs time
```

```python
plt.xlabel('time')  # create xlabel
plt.ylabel('position')  # create ylabel
```

```python
# statements for the same plot need to be in the same cell
plt.plot(time, position)
plt.xlabel('time')
plt.ylabel('position')
```

```python
# plotting from pandas dataframes
import pandas as pd
data = pd.read_csv('data/gapminder_gdp_oceania.csv', index_col = 'country')
```

```python
data.plot()   # use plot() function directly with DataFrame, one line per column
```

```python
data.T.plot()  # transpose data before plotting -> one line per country
```

```python
# change the column names in data, so that they are properly used as labels for x-axis
# 1. extract year from last 4 characters of each column name
years = data.columns.str.strip('gdpPercap_')   # str.strip to strip off the part in ' '
print(years)
```

```python
# replace column names in data with variable years and change type of years to integer
data.columns = years.astype(int)
```

```python
data.T.plot()
```

## Exporting a plot

```python
# having the savefig function in a cell without the plot itself only saves an empty plot
plt.savefig('myplot.png')
```

```python
# put plt.savefig after plot statements, but before plot is printed in notebook
data.T.plot()
plt.savefig('myplot.png')
```
